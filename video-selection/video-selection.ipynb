### Videos each channel across the last 6 months

```python
import json
import time

result_json = {}
request_counter = 0

# setup API key
api_key = 'AIzaSyA8jJaGJKRs4ImuDeHsH458dr4AU0ai7Jc'

def request_videos(name, category):
    
    '''Function takes a dictionary of channels with their ID from a category
    and outputs the videos of that channel across the last 6 months (max 50)'''
    
    videos = {}
    
    for j in category:
        # request body
        request = youtube.search().list(
                part='snippet, id',
                channelId=category[j],
                order='date',
                publishedAfter='2020-11-14T00:00:00Z',
                maxResults=50
            )
        # execute request to get channel videos
        try:
            response = request.execute()
            for i in response['items']:
                video_id = i['id']['videoId']
                # request body for single video
                request_views = youtube.videos().list(
                        part='statistics',
                        id=video_id
                    )
                # execute request and save title, published date, views and URL
                response_views = request_views.execute()
                views = response_views['items'][0]['statistics']['viewCount']
                newlist = []
                newlist.append(i['snippet']['title'])
                newlist.append(i['snippet']['publishedAt'])
                newlist.append(views)
                newlist.append('http://www.youtube.com/watch?v=' + video_id)
                try:
                    # new entry in dictionary
                    videos[j].append(newlist)
                except:
                    # add infos of video to already existing entry
                    videos[j] = [newlist]
        except:
            continue
            
    # add dictionary of one category to overall dictionary        
    result_json[name] = videos
    
    # uncomment this for separate json files
    # timestring = time.strftime("%Y-%m-%d_%H:%M:%S")
    # filename = f"SC_Research_Project_{name}_Data_{timestring}.json"
    # with open(filename, 'w') as file:
        # file.write(json.dumps(result_json, indent=2))
    

# method call for each category

# setup API with key
api_key = 'AIzaSyA8jJaGJKRs4ImuDeHsH458dr4AU0ai7Jc'
youtube = build('youtube', 'v3', developerKey=api_key)

request_videos("Gaming", gaming_id)
request_videos("Comedy", comedy_id)
request_videos("Sports", sports_id)

# setup API with key
api_key = 'AIzaSyA8sJ2JRK7cd8sgmBX9r3p6ERht2XLuFgM'
youtube = build('youtube', 'v3', developerKey=api_key)

request_videos("Kids", kids_id)
request_videos("How-To", howto_id)
request_videos("Pets", pets_id)
request_videos("Science", science_id)

# stringify dictionary to JSON string
print(json.dumps(result_json, indent=2))

# save contents to file with timestamp
timestring = time.strftime("%Y-%m-%d_%H:%M:%S")
filename = f"SC_Research_Project_Videos_Selection_Data_{timestring}.json"
with open(filename, 'w') as file:
    file.write(json.dumps(result_json, indent=2))
    
# number requests
print(f"\nrequests for video information: {request_counter}")
print(f"total requests (incl. requests for channel IDs): {total_request_counter}")
```
