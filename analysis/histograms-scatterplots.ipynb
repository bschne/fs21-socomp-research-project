```python
import numpy as np

def plot(data, category, metric, x_label):
    
    data = data[data.between(data.quantile(.15), data.quantile(.85))]

    x = []
    counter = 0
    
    for i in data:
        if i != 0:
            x.append(i*100000)
        else:
            counter+=1
        

    
    
    fig, ax = plt.subplots(figsize=(18, 10))

    ax.set_xlabel(x_label)
    ax.set_title(category +' '+ metric +' | Amount of 0 '+ metric +': '+ str(counter))
    ax.set_ylabel('Number of videos')

    x_scale = 5
    y_scale = 20
    plt.hist(x, 120)
    plt.xlim(0, x_scale)
    plt.ylim(0, y_scale)

    plt.savefig(category +' '+ metric +' '+ str(x_scale) +'x'+ str(y_scale))
    plt.show()
    print(counter)
```
